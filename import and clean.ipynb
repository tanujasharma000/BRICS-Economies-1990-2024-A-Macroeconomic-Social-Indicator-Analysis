```python


import wbgapi as wb
import pandas as pd
from functools import reduce

brics = ['BRA', 'RUS', 'IND', 'CHN', 'ZAF']

indicators = {
    # ----- ECONOMY (Economy) -----
    'NY.GDP.MKTP.CD': 'GDP_Current_USD',                     # Total GDP (current $)
    'NY.GDP.PCAP.CD': 'GDP_per_Capita_USD',                  # Per capita GDP
    'NY.GDP.MKTP.KD.ZG': 'GDP_Growth_Annual_Pct',            # GDP growth rate
    'NY.GDP.MKTP.PP.CD': 'GDP_PPP',                          # Purchasing Power Parity
    'TX.VAL.MRCH.CD.WT': 'Total_Exports_USD',                # Exports
    'TM.VAL.MRCH.CD.WT': 'Total_Imports_USD',                # Imports
    'NE.CON.TOTL.KD.ZG': 'Household_Consumption_Growth',     # Consumption

    # ----- POPULATION & DEMOGRAPHICS -----
    'SP.POP.TOTL': 'Total_Population',
    'SP.POP.GROW': 'Population_Growth_Pct',
    'SP.URB.TOTL.IN.ZS': 'Urban_Population_Pct',
    'SP.DYN.LE00.IN': 'Life_Expectancy_Years',               # Health indicator

    # ----- HEALTH & EDUCATION -----
    'SH.DYN.MORT': 'Under5_Mortality_Rate',                  # Child mortality
    'SE.ADT.LITR.ZS': 'Literacy_Rate_Adult_Pct',
    'SE.XPD.TOTL.GD.ZS': 'Govt_Education_Expenditure_Pct_GDP',

    # ----- LABOR & EMPLOYMENT -----
    'SL.UEM.TOTL.ZS': 'Unemployment_Pct',
    'SL.TLF.CACT.ZS': 'Labor_Force_Participation_Rate',

    # ----- ENVIRONMENT & ENERGY -----
    'EN.ATM.CO2E.PC': 'CO2_Emissions_Per_Capita',
    'EG.FEC.RNEW.ZS': 'Renewable_Energy_Consumption_Pct',

    # ----- TECHNOLOGY & INFRASTRUCTURE -----
    'IT.NET.USER.ZS': 'Internet_Users_Pct_Population',
    'IT.CEL.SETS': 'Mobile_Subscriptions',
}

print(f" Downloading data for {len(indicators)} indicators from 1960 to 2025...")

# Sabhi indicators ka data fetch karke list mein store karein
all_dfs = []

for code, name in indicators.items():
    print(f"   Fetching: {name} ...")
    try:
        
        df = wb.data.DataFrame(code, brics, range(1960, 2025))
        
        df = df.reset_index().melt(id_vars=['economy'], var_name='Year', value_name=name)
        all_dfs.append(df)
    except Exception as e:
        print(f"   ❌ Error in {name}: {e}")

print("\n Merging all indicators into one dataset...")
df_final = reduce(lambda left, right: pd.merge(left, right, on=['economy', 'Year'], how='outer'), all_dfs)

df_final.to_csv('BRICS_Complete_Analysis_Dataset.csv', index=False)

print("\n SUCCESS! Dataset saved as 'BRICS_Complete_Analysis_Dataset.csv'")
print(f" Total Rows: {df_final.shape[0]}, Total Columns: {df_final.shape[1]}")
print(df_final.head())
```

```python
df=pd.read_csv('BRICS_Complete_Analysis_Dataset.csv')
missing_per_year=df.groupby('Year').apply(lambda x:x.isnull().sum().sum())
print(missing_per_year.head(40))
```

```python
df_main = df_main.drop(
    columns=[
        'Literacy_Rate_Adult_Pct',
        'Govt_Education_Expenditure_Pct_GDP'
    ]
)
```

```python
df_main=df_main.sort_values(['economy','Year'])
numeric_cols=df_main.select_dtypes(include='number').columns
df_main[numeric_cols]=(df_main.groupby('economy')[numeric_cols].transform(lambda x:x.interpolate()))
df_main[numeric_cols]=df_main.groupby('economy')[numeric_cols].transform(lambda x: x.ffill().bfill())
```
